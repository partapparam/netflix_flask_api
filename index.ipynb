```python
import pandas as pd

```

```python
df = pd.read_csv('./netflix_content.csv')
df.head()
```

```python
clean_df = df.drop('index', axis=1)
clean_df
```

```python
# rename 'id' col to 'netflix id'
# {“old_column_name” : “new_column_name”}
clean_df.rename(columns={'id' : 'netflix_id'}, inplace=True)
cols = clean_df.columns
```

```python
# comma separate the DF column names
cols = ','.join(list(clean_df.columns))
# create a list of tuples from the dataframe values
tuples = [tuple(x) for x in clean_df.to_numpy()]
query = 'INSERT INTO %s(%s) VALUES (%%s, %%s, %%s, %%s,%%s, %%s,%%s, %%s,%%s, %%s)' % ('content', cols)

```

```python
import psycopg2
conn = psycopg2.connect(database='movies', user='param')
```

```python
cursor = conn.cursor()
cursor.executemany(query, tuples)
```
